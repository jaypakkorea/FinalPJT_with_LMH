영화 1 : 영화 평점이 10/10 (5명 평가)
영화 2 : 영화 평점이 8/10 (500명 평가) 
영화 2가 더 신뢰도 높음
따라서 평점 정보를 바로 쓸 수 없고 가중치가 필요함

```python
# 영화 전체 평점 평균

C = df2['vote_average'].mean()
C
```

```python
# m 즉 너무 적은 데이터는 버린다
# 0.9로 상위 10% 평가 받은 영화들의 수 ,

m = df2['vote_count'].quantile(0.9)
m
```

```python
# 상위 10%에 드는 데이터만 따로 뽑아서 q_movies라는 데이터를 다시 만듬

q_movies = df2.copy().loc[df2['vote_count'] >=m]
q_movies
```

```python
# 왼쪽이 영화 id, 오른쪽이 평가를 몇 번 받았는지
# 가장 적게 받은 평가가 1840개, 가장 많이 받은 영화는 13752개 받았다는 뜻
q_movies['vote_count'].sort_values()
```

```python
# IMDB에서 제공한 평가 시스템을 기반으로 가중치를 계산하는 함수를 만든다

def weighted_rating(x,m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v+m) * R) + ((m+v) * C)
```

```python
# score 컬럼을 하나 더 만들어 가중치를 계산한 값은 넣어준다

q_movies['score'] = q_movies.apply(weighted_rating, axis = 1)
q_movies.head(3)
```

```python
#가중치를 기준으로 가장 높은 10개의 제목, 투표 수, 평균점수, 가중치

q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title_x', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
# popularity 는 갑자기 조회가 급증하는 영화
# and is also just kind of fun to see items chart up and down.

pop= df2.sort_values('popularity', ascending=False)
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))

plt.barh(pop['title_x'].head(10),pop['popularity'].head(10), align='center',
        color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel("Popularity")
plt.title("Popular Movies")
```

```python
indices = pd.Series(df2.index, index=df2['title_x']).drop_duplicates()
indices
```

```python
#제목을 입력받으면 가장 유사도가 높은 10개의 영화 추출

def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x : x[1], reverse = True)
    sim_scores[1:11]
    movie_indices = [i [0] for i in sim_scores ]
    return df2['title_x'].iloc[movie_indices]
```

```python
# 3개만 가져오자 

def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names
    #Return empty list in case of missing/malformed data
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title_x', 'cast', 'director', 'keywords', 'genres' ]].head(3)
```

```python
# 대문자 소문자를 하나로 합치기 (vector화)

def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    else :
        if isinstance(x, str):
            return str.lower(x.replace(' ', ''))
        else :
            return ''
                            
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features :
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title_x', 'cast', 'director', 'keywords', 'genres' ]].head(3)
```

```python
# 콤마 (,)를 지우고 단어들로 나열하는 함수
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```
